## Prewitt filters

**<a id="#">Task 4.1.1.</a>** Load the image `data/astronaut.png` and show it:

**<a id="#">Task 4.1.2.</a>** Implement horizontal and vertical 3&times;3 Prewitt filters:

```python
def prewitt_h(img):
    # write your code here

    # Berechnet die horizontale Ableitung g_x mit dem Prewitt-Operator.
    result = np.zeros_like(img, dtype=np.float32)
    kernel = np.array([[1, 0, -1], 
                       [1, 0, -1], 
                       [1, 0, -1]])
    
    n = 1  # FiltergrÃ¶ÃŸe 3x3 â†’ Rand von 1 Pixel
    
    for row in range(n, img.shape[0] - n):
        for col in range(n, img.shape[1] - n):
            neighborhood = img[row - n : row + n + 1, col - n : col + n + 1]
            result[row, col] = np.sum(neighborhood * kernel)
    
    return result

g_x = prewitt_h(img)
    
def prewitt_v(img):
    # write your code here

    """Berechnet die vertikale Ableitung g_y mit dem Prewitt-Operator."""
    result = np.zeros_like(img, dtype=np.float32)
    kernel = np.array([[1, 1, 1], 
                       [0, 0, 0], 
                       [-1, -1, -1]])
    
    n = 1  # FiltergrÃ¶ÃŸe 3x3 â†’ Rand von 1 Pixel
    
    for row in range(n, img.shape[0] - n):
        for col in range(n, img.shape[1] - n):
            neighborhood = img[row - n : row + n + 1, col - n : col + n + 1]
            result[row, col] = np.sum(neighborhood * kernel)
    
    return result

g_y = prewitt_v(img)
```
